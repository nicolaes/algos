(BASE * min(N/500, 1))^((N-C+1)/N)

sau BASE^((N-C+1)/N)

```python

from math import ceil
import matplotlib.pyplot as plt

BASE = 2500
N = 5750

get_cps = lambda c: ceil((BASE * min(N/500, 1)) ** ((N-c+1)/N))
ranks = range(1, 3000, 100)
points = [get_cps(c) for c in ranks]

plt.plot(ranks, points)
plt.show()
```
